Assignment : Hospital Patient Management System

A hospital is struggling to manage its patient records, doctors’ schedules, billing, and appointments using manual methods. They want to develop a Python-based system using Object-Oriented Programming (OOP) principles to handle the following:

1.Patient Management
* Each patient has a unique ID, name, age, gender, contact info, and medical history.
* Patients should be able to book appointments with available doctors.
* The system should maintain a record of a patient’s past visits, prescribed medicines, and lab tests.

2.Doctor Management
* Doctors have a unique ID, name, specialization, and schedule (days/times available).
* The system should allow assigning doctors to patients based on specialization and availability.

3.Appointment Scheduling
* Patients should be able to request appointments.
* The system should check the doctor’s availability and confirm or suggest alternate slots.

4.Billing After a consultation, generate bills that include doctor’s fee, lab test charges (if applicable), and   medicines (if applicable). Bills should be linked to patient records.

5.Reports Number of patients seen by each doctor (Overall count) Number of patients seen by each doctor per month Number of patients seen by each doctor per day Total amount paid by specific patient (Overall)

Inheritance: Inpatient and Outpatient both extend Patient class.

Polymorphism: Each class has its own info() method; Python calls the correct one at runtime.

Encapsulation: Encapsulation combines the data an object holds and the functions that manipulate that data into a single, cohesive unit : the class , applicable in all 3 classes here

```python
class Patient:
  def __init__(self,patient_id,name,age,gender,contact):
    self.patient_id = patient_id
    self.name = name
    self.age = age
    self.gender = gender
    self.contact = contact
    self.__medical_history = []

  def info(self):
    return f"Patient-ID : {self.patient_id}, Name : {self.name}, Age : {self.age}, Gender : {self.gender}, Contact : {self.contact}, Medical History : {self.__medical_history}"

  def add_medical_history(self,record):
    return self.__medical_history.append(record)

  def get_medical_history(self):
    return self.__medical_history
```

```python
class inpatient(Patient):
  def __init__(self,patient_id,name,age,gender,contact,room_no):
     super().__init__(patient_id,name,age,gender,contact)
     self.room_no = room_no

  def info(self):
    return f"Patient-ID : {self.patient_id}, Name : {self.name}, Age : {self.age}, Gender : {self.gender}, Contact : {self.contact}, Room-no : {self.room_no}"


class Outpatient(Patient):
  def __init__(self,patient_id,name,age,gender,contact,visit_date):
    super().__init__(patient_id,name,age,gender,contact)
    self.visit_date = visit_date

  def info(self):
    return f"Patient-ID : {self.patient_id}, Name : {self.name}, Age : {self.age}, Gender : {self.gender}, Contact : {self.contact}, Visit-date : {self.visit_date}"
```

```python
class Doctor:
  def __init__(self,doctor_id,name,specialization):
    self.doctor_id = doctor_id
    self.name = name
    self.specialization = specialization
    self.schedule = {}

  def add_avaliability(self,date, time_slot):
      self.schedule[date] = time_slot

  def info(self):
      return f"Doctor-ID :{self.doctor_id}, Name :{self.name}, Specialization : {self.specialization}, Schedule : {self.schedule}"
```

```python
class Appointment:
  def __init__(self,appointment_id,patient,doctor,date,time):
    self.appointment_id = appointment_id
    self.patient = patient
    self.doctor = doctor
    self.date = date
    self.time = time

  def info(self):
   return f"Appointment-ID :{self.appointment_id}, Patient Name : {self.patient.name} with Dr.{self.doctor.name} on {self.date} at {self.time}"
```

```python
class Bill:
  def __init__(self,bill_id,patient,doctor_fee,lab_charges=0,medicines=0):
    self.bill_id = bill_id
    self.patient = patient
    try:
      self.__doctor_fee = int(doctor_fee)
      self.__lab_charges = int(lab_charges)
      self.__medicines = int(medicines)
    except ValueError:
      print("Bill should be in Numbers")
      self.__doctor_fee = 0
      self.__lab_charges = 0
      self.__medicines = 0

  def calc_total(self):
    return self.__doctor_fee + self.__lab_charges + self.__medicines

  def info(self):
    return f"Bill No : {self.bill_id}, Patient name : {self.patient.name}, Doctor Fees : {self.__doctor_fee}, Lab Charges : {self.__lab_charges}, Medicines : {self.__medicines} \nTotal Amount : {self.calc_total()}"
```

```python
bill2 = Bill("002",p4,600,150,1000)
print(bill2.info())
```

```python
#5.Reports Number of patients seen by each doctor (Overall count) Number of patients seen by each doctor per month Number of patients seen by each doctor per day Total amount paid by specific patient (Overall)
class Hospital:
  def __init__(self):
    self.appointments = []
    self.bills = []

  def book_appointment(self, appointment):
    try:
      date = appointment.date
      time = appointment.time
      if date not in appointment.doctor.schedule or time not in appointment.doctor.schedule[date]:
        raise ValueError(f"Dr. {appointment.doctor.name} is not avaliable on {date} at {time}.")

      self.appointments.append(appointment)
      print("Confirmed Appointment :-", appointment.info())

    except ValueError as error_message:
      print(f"Booking Falied : {error_message}")

  def generate_bill(self, bill):
    self.bills.append(bill)
    print("Bill Generated :-", bill.info())

  def No_ofpatientsseenby_doctor(self,doctor_name):
    count = sum(1 for app in self.appointments if app.doctor.name == doctor_name)
    print(f"Dr. {doctor_name} has seen {count} patients averall.")

  def No_ofpatientsseenbydoctor_eachmonth(self, doctor_name, date):
    count = sum(1 for app in self.appointments if app.doctor.name == doctor_name and app.date == date)
    print(f"Dr. {doctor_name} has seen {count} patients on {date}.")

  def Total_amountpaidbyspecific_patient(self, patient_name):
    total = sum(bill.calc_total() for bill in self.bills if bill.patient.name == patient_name)
    print(f"Total Amount Paid By {patient_name} is Rs {total}.")
```
