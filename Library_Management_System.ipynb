**Library Management Systems**

The Goal: A system where you can add, update, delete, and search for items (books or products).

```python
class Person:
  def __init__(self, person_id,name,contact):
    self.person_id = person_id
    self.name = name
    self.contact = contact
    self.__taken_books = []

  def info(self):
    return f"Person ID : {self.person_id}, Name : {self.name}, Contact : {self.contact}, Taken Books : {self.__taken_books}"

  def add_taken_books(self, book_name):
    return self.__taken_books.append(book_name)

  def get_taken_books(self):
    return self.__taken_books
```

```python
p.add_taken_books("Goast Life")
```

```python
p.get_taken_books()
p.info()
```

```python
class Student(Person):
  def __init__(self,person_id,name,contact,standard):
     super().__init__(person_id,name,contact)
     self.standard = standard

  def info(self):
     return f"Person ID : {self.person_id}, Name : {self.name}, Contact : {self.contact}, Standard : {self.standard}"

class Faculty(Person):
  def __init__(self,person_id,name,contact,department):
     super().__init__(person_id,name,contact)
     self.department = department

  def info(self):
      return f"Person ID : {self.person_id}, Name : {self.name}, Contact : {self.contact}, Department : {self.department}"
```

```python
f1 = Faculty(1002,"Manoj",4789025410,"Science")
f1.info()
```

```python
class Book:
  def __init__(self,book_id,title,author,quantity):
    self.book_id = book_id
    self.title = title
    self.author = author
    self.__avaliable_books = quantity

  def check_books(self):
    return self.__avaliable_books

  def remaining_books(self):
    self.__avaliable_books -= 1

  def info(self):
    return f"Book-ID : {self.book_id}, Title : {self.title} by Author : {self.author}, Remaining : {self.__avaliable_books}"
```

```python
b1 = Book("101-A","Spider-Man","Les-wein",2)
b1.info()
```

```python
b2 = Book(502, "Captian-America", "Ron Goulart", 0)
b2.info()
```

```python
class Transaction:
  def __init__(self,Transaction_id, person,book, date):
    self.Transaction_id = Transaction_id
    self.person = person
    self.book = book
    self.date = date

  def info(self):
    return f"Transaction ID {self.Transaction_id} : {self.book.title} issued to {self.person.name} on {self.date}"
```

```python
#Invalid Transaction
t3 = Transaction("T003",s1,b2,"22-05-2026")
t3.info()
```

```python
class Fine:
  def __init__(self,fine_id, person,days_left):
     self.fine_id = fine_id
     self.person = person
     self.fee_perday = 10
     try:
      self.__days_left = int(days_left)
     except ValueError:
       print(f"ERROR in Fine {self.fine_id}: Days late must be a number.")
      #  self.__days_left = 0

  def cal_fine(self):
    return self.__days_left * self.fee_perday

  def info(self):
    return f"Fine No - {self.fine_id} ,Name : {self.person.name}, Days Left : {self.__days_left} \nTotal Fine : {self.cal_fine()}"
```

```python
class Library:
  def __init__(self):
    self.Transactions = []
    self.fines = []

  def issue_book(self, Transaction):
    try:
      target_book = Transaction.book
      if target_book.check_books() <= 0:
        raise ValueError(f"Sorry, {target_book.title} is currently out of stock")

      target_book.remaining_books()
      Transaction.person.add_taken_books(target_book.title)
      self.Transactions.append(Transaction)
      print("Successfully Issued : ", Transaction.info())

    except ValueError as error_message:
      print(f"Issued Failed : ", {error_message})

  def collect_fine(self,fine):
    self.fines.append(fine)
    print("Generated Fine : " + fine.info())

  def report_books_by_person(self,person_name):
    count = sum(1 for trans in self.Transactions if trans.person.name == person_name)
    print(f"Report : {person_name} has borrowed {count} books averall.")

  def report_book_popularity(self,book_title):
    count = sum(1 for trans in self.Transactions if trans.book.title == book_title)
    print(f"Report : The Book {book_title} has been issued {count} times.")
```

```python
patshala = Library()
```

```python
#valid transactions
t1 = Transaction("T001",s1,b1, "15-05-2026")
patshala.issue_book(t1)
```

```python
#invalid transaction
t2 = Transaction("T002",f1,b2,"25-05-2026")
patshala.issue_book(t2)
```

```python
t3 = Transaction("T003",f1,b1,"22-05-2026")
patshala.issue_book(t3)
```
